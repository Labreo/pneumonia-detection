```python
%matplotlib inline
import matplotlib.pyplot as plt
import os
root_dir = os.getcwd()  # get current working directory
src_path = os.path.join(root_dir, "../data/chest_xray/train/")
print(os.path.exists(src_path))

sub_class= os.listdir(src_path)
#Removing so that it doesnt cause issue when plotting unnecressary macosx file just ignore
sub_class=[folder for folder in os.listdir(src_path) if folder !='.DS_Store']
fig = plt.figure(figsize=(10,5))
#lets visualize
path = os.path.join(src_path,sub_class[0])
for class_index, class_name in enumerate(sub_class):
    path = os.path.join(src_path, class_name)

    #  Get first 4 image files with valid extensions
    img_files = [f for f in os.listdir(path) if f.endswith(('.jpg', '.jpeg', '.png'))][:4]

    for i in range(4):
        img_path = os.path.join(path, img_files[i])
        img = plt.imread(img_path)

        plt.subplot(2, 4, class_index * 4 + i + 1)
        plt.imshow(img, cmap='gray')
        plt.title(class_name)
        plt.axis('off')

plt.tight_layout()
plt.show()

```
